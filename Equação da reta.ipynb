```python
# Coordenadas do ponto

ponto = []
for i in range(1, 4):
    I = input('Insira a coordenada do ponto: ')
    ponto.append(int(I))
    
vetor = []   
for i in range(1, 4):
    I = input('Insira a coordenada do vetor: ')
    vetor.append(int(I))

PX = int(ponto[0])
PY = int(ponto[1])
PZ = int(ponto[2])

VX = int(vetor[0])
VY = int(vetor[1])
VZ = int(vetor[2])
```

```python
# Importando as bibliotecas

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
#Mostrando graficamente os pontos no grfico 3d


x = [PX, VX]
y = [PY, VY]
z = [PZ, VZ]
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
ax.scatter3D(x, y, z, color='red')
plt.legend("P")
plt.show()
```

```python
# Mostrando a reta formada pelo ponto e o vetor
fig2 = plt.figure()
grafico = fig2.add_subplot(projection='3d')



X = np.linspace(PX,VX,2)
Y = np.linspace(PY,VY,2)
Z = np.linspace(PZ,VZ,2)
grafico.plot(X,Y,Z)
plt.title("Reta formada pelo ponto e vetor")
plt.show()
```

```python
#Para descobrir se o ponto pertence a reta

pontox = []   
for i in range(1, 4):
    I = input('Insira a coordenada do ponto desconhecido: ')
    pontox.append(float(I))
```

```python
px = float(pontox[0])
py = float(pontox[1])
pz = float(pontox[2])
```

```python
eixox = (px - PX)/VX
eixoy = (py - PY)/VY
eixoz = (pz - PZ)/VZ
```

```python
if ((eixox == eixoy) & (eixoy == eixox)):
    print("Pertence a mesma reta")
```

```python
fig2 = plt.figure()
grafico = fig2.add_subplot(projection='3d')



X = np.linspace(PX,px,2)
Y = np.linspace(PY,py,2)
Z = np.linspace(PZ,pz,2)
grafico.plot(X,Y,Z)
plt.title("Reta formada pelo ponto desconhecido")
plt.show()
```
